Overfitting can be caused by trees which have a large depth, too many decision nodes, or too many leaf nodes. In this case, a solution very close to, but not at, the global optimum is reached, with a test accuracy score of 0.95. This is a cause of the decision nodes being locally greedy, meaning that binary splits are made without the consideration of the entire solution. Consequently, a local optima is achieved.

```python
#implementation of decision tree
#import packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df = pd.read_csv('CS51_Assignment_Diagnosis_Data (1).csv') #import data

all_inputs = df[['Temperature', 'WBC Count', 'Headache Severity', 'Cough Severity']].values #array of values
all_classes = df[ 'Infected'].values #array of values

def accuracy(n): #simulations to calculate averages for training and testing accuracies with different data splits
    train_array = [] #for training accuracy values
    test_array = [] #for testing acccuracy values
    for i in range(n): #loop to run simulations
        (train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, test_size=0.3,random_state=1)
        #randomise data into training and testing groups 
        dtc = DecisionTreeClassifier(max_depth = 3, min_samples_leaf = 1) #create boundaries to split data in decision tree
        dtc.fit(train_inputs, train_classes) #fit training data into decision tree leaf nodes
        train_array.append(dtc.score(train_inputs, train_classes)) #append accuracy of tree for training data
        test_array.append(dtc.score(test_inputs, test_classes)) #append accuracy of tree for testing data
    train_avg = sum(train_array) / n #avg training accuracy
    test_avg = sum(test_array) / n #avg testing accuracy
    return train_avg, test_avg #output averages

average_accuracies = accuracy(1000) #call function to run simulations
print("Training Accuracy score:", average_accuracies[0]) #print training accuracy
print("Test Accuracy score:", average_accuracies[1]) #print testing accuracy
plt.figure(figsize = (15, 10)) #increase figure size 
tree.plot_tree(dtc, feature_names = ['Temperature', 'WBC Count', 'Headache Severity', 'Cough Severity'], class_names = ['Not infected', 'Infected'], rounded = True) #plot tree
```

#### 2.1.2 Eulerâ€™s Method Description (~150 words) [#algorithms]

Explain what it means to solve the SIR differential equations and how Eulerâ€™s method works as an algorithm to achieve a numerical solution via simulation. In your explanation, identify whether the model and simulation are continuous or discrete, and consider the role of the step size (`h`) in the algorithm.

To solve the SIR equations means to determine the set of mathematical functions S(t), I(t), R(t) which allow us to see the numbers of individuals who are susceptible, infected and recovered from Smallpox over time, based on the preset parameters of infection and recovery rates, b and k. 

To model the spread of Smallpox, I have used two extra parameters, d and v, to produce differential equations to model the deaths from Smallpox, as well as the number of vaccinated individuals over time after the vaccine has been produced:

$$ \frac{dS}{dt} = -\frac{b}{N}S(t) I(t) - v S(t) $$
<br>
$$ \frac{dI}{dt} = \frac{b}{N}S(t)I(t) - kI(t) $$
<br>
$$ \frac{dR}{dt} = (1-d) k I(t) $$
<br>
$$ \frac{dD}{dt} = d k I(t) $$
<br>
$$ \frac{dV}{dt} = v S(t) $$
<br>

Euler's method allows us to numerically solve these equations, to produce graphs of S, I, R, D, and V over time through approximations. Using an initial point with the given differential equations for the SIR model, subsequent points can be predicted, each using the co-ordinates of the previous point and the tangent at the given x-value to find the next point.

Euler's method provides a discrete approximation to a continuous-time system as the state variables, S, I, R, D, and V only change at a set number of points over time.

Smaller step sizes (h) can be used to increase the accuracy of the model by reducing the aggregate error which increases with each additional point in the model. However, this increases the computer power required to find the solution, therefore it can be necessary to find a balance between these two factors. The use of Euler's method allows us to find a very close approximation to the analytical solution by using a small step value, and without having to solve differential equations.

#### 2.1.3 Eulerâ€™s Method Implementation [#algorithms, #dataviz]

Implement a numerical simulation to derive the implications of your model using Eulerâ€™s method in Python.Your simulation must generate at least one relevant visualization of the disease dynamics, including a descriptive figure legend and caption. You may need to adjust the run-time and step size in your simulation to ensure the visualization is maximally informative. Include thorough comments in your code to convey your understanding of the implementation of Eulerâ€™s method. 

```python
#import packages
import numpy as np 
import matplotlib.pyplot as plt

t0 = 0 #starting value for t
t1 = 120 #ending value for t

h = 1 #step size
steps = int((t1 - t0)/h + 1) #number of steps

t = np.linspace(t0, t1, steps) #storing t values
S = np.zeros(steps) #for storing S values
I = np.zeros(steps) #for storing I values
R = np.zeros(steps) #for storing R values
D = np.zeros(steps) #for storing D values
V = np.zeros(steps) #for storing V values

b = 0.5 #infection rate
k = 0.15 #recovery rate
d = 0.3 #death rate (Signs and Symptoms, 2021)
v = 0.05 #vaccination rate, 5% (number of individuals vaccinated per unit t)

def dSdt(t, S, I): #function for susceptible individuals 
    if t > 40: #after vaccine
        return (-b / N) * S * I - v * S #return number of individuals susceptible at time t after vaccine
    else: #before vaccine
        return (-b / N) * S * I #return number of individuals susceptible at time t before vaccine

def dIdt(t, S, I): #function for infected individual 
    return ((b / N) * S * I) - (k * I) #return number of individuals infected at time t

def dRdt(t, I): #function for recovered individuals
    return (1 - d) * (k * I) #return number of individuals recovered at time t

def dDdt(t, I): #function for dead individuals
    return d * (k * I) #return number of dead individuals at time t

def dVdt(t, S): #function for vaccinated individuals
    if t > 40: #after vaccine
        return v * S #return number of vaccinated individuals per day
    else: #before vaccine
        return 0 #none vaccinated before vaccine invented

S[0] = 56180000 #initial susceptible individuals, population of UK (United Kingdom  - Place Explorer - Data Commons, 2021)
I[0] = 1 #initial infected individuals, initial carrier of Smallpox to the UK
R[0] = 0 #initial recovered individuals
D[0] = 0 #intitial deceased individuals 
V[0] = 0 #initial vaccinated individuals
N = S[0] + I[0] + R[0] + D[0] + V[0] #total individuals in system = sum of variables

for n in range(steps-1): #steps through points in time 
    S[n+1] = S[n] + dSdt(t[n],S[n],I[n]) #adds next cumulative susceptible count to array
    I[n+1] = I[n] + dIdt(t[n],S[n],I[n]) #adds next cumulative infected count to array
    R[n+1] = R[n] + dRdt(t[n], I[n]) #adds next cumulative recovered count to array
    D[n+1] = D[n] + dDdt(t[n], I[n]) #adds next cumulative dead count to array
    V[n+1] = V[n] + dVdt(t[n], S[n]) #adds next cumulative vaccinated count to array

plt.rcParams.update({'font.size': 15}) #set font size
plt.rcParams["figure.figsize"] = [8,5] #set figure size
plt.plot(t,S,linewidth=2,label='S(t)') #plot cumulative susceptible over time
plt.plot(t,I,linewidth=2,label='I(t)') #plot cumulative infected over time
plt.plot(t,R,linewidth=2,label='R(t)') #plot cumulative recovered over time
plt.plot(t,D,linewidth=2,label='D(t)') #plot cumulative dead over time
plt.plot(t,V,linewidth=2,label='V(t)') #plot cumulative vaccinated over time
plt.xlabel('Time (days)', fontsize = 8) #label x axis
plt.ylabel('Number of individuals (10s of millions)', fontsize = 8) #label y axis
plt.title('Number of Individuals in S, I, R, D, and V Over Time', fontsize = 14) #set title
plt.legend(loc = 'best') #place legend on graph in optimal place
plt.show() #display graph
```

```python
#import packages
import networkx as nx
import EoN
import matplotlib.pyplot as plt
from collections import defaultdict

G = nx.grid_2d_graph(100,100) #each node is (u,v) where 0<=u,v<=99, creates 100x100 grid of agents
initial_infections = [(u,v) for (u,v) in G if 45<u<55 and 45<v<55] #agents that are initially infected

H = nx.DiGraph() #the spontaneous transitions
H.add_edge('Sus', 'Vac', rate = 0.01) #vaccination rate
H.add_edge('Inf', 'Rec', rate = 1.0) #rate of individuals that will recover from being infected

J = nx.DiGraph() #the induced transitions
J.add_edge(('Inf', 'Sus'), ('Inf', 'Inf'), rate = 2.0) #rate of people that become infected from susceptible

IC = defaultdict(lambda:'Sus') #set all initial agents (co-ordinates) to susceptible
for node in initial_infections: #loop through initially infected nodes
    IC[node] = 'Inf' #set initial infected nodes to infected from susceptible
    
return_statuses = ['Sus', 'Inf', 'Rec', 'Vac'] #determine numbers of each group

color_dict = {'Sus': 'white','Inf':'red', 'Rec':'green','Vac': 'blue'} #change colours of agents
pos = {node:node for node in G} #cycles through grid of agents for co-ordinates
tex = False #eliminate italics
sim_kwargs = {'color_dict':color_dict, 'pos':pos, 'tex':tex} #set simulation parameters

sim = EoN.Gillespie_simple_contagion(G, H, J, IC, return_statuses, tmax=30,
return_full_data=True, sim_kwargs=sim_kwargs) #run simulation returning data, using set parameters

sim.display(time=8, node_size = 4, ts_plots=[['Inf']]) #set display parameters and variables (time run, size of agents, plots to make)
plt.xlabel('Time (days)') #set x axis label
plt.ylabel('Number of individuals') #set y axis label
plt.title('Number of Infected Individuals Over Time') #set title
plt.tight_layout() #to prevent overlapping of figures
plt.show() #display figures
```
